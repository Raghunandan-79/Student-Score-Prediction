```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load dataset
data = pd.read_csv("student_performance.csv")
```

```python
# input and output
x = data[['weekly_self_study_hours']]
y = data['total_score']
```

```python
# train model
model = LinearRegression()
model.fit(x, y)
predicted_scores = model.predict(x)
```

```python
# valid regression metrics
mae = mean_absolute_error(y, predicted_scores)
mse = mean_squared_error(y, predicted_scores)
rmse = np.sqrt(mse)
r2 = r2_score(y, predicted_scores)

# show results
print("Mean Absolute ERROR (MAE): ", round(mae, 2))
print("Mean Squared ERROR (MSE): ", round(mse, 2))
print("Root Mean Squared ERROR (RMSE): ", round(rmse, 2))
print("R^2 Score (Model Accuracy): ", round(r2, 4)) # closer to 1 = better
```

```python
# histogram
plt.figure(figsize=(10, 6))
plt.hist(data["total_score"], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Final Exam Scores")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.tight_layout()
plt.savefig("Distribution of Final Exam Scores.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Scatter + Regression Line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color="blue", label="Actual Scores")
plt.plot(x, predicted_scores, color="red", label="Predicted Scores (Regression Line)")
plt.title("Model Prediction VS Actual Score")
plt.xlabel("Study Hours Per Week")
plt.ylabel("Final Output")
plt.grid(True)
plt.tight_layout()
plt.savefig("Model Prediction VS Actual Score.png", dpi=300, bbox_inches="tight")
plt.show()
```
